# BBO Challenge Round 5: Architectural Refinement

This notebook implements a hierarchical refinement strategy for the Black-Box Optimisation (BBO) challenge.

**Approach:** Start from the best known region (centre of the search space) and apply small, normally-distributed perturbations. The reduced step size (`0.05` vs the `0.1` learning rate in Round 4) reflects a shift from broad exploration to fine-grained exploitation, simulating the refinement phase of a surrogate-guided search.

## Setup & Configuration

```python
import numpy as np

# Set seed for Round 5 / Module 17 context
np.random.seed(171)
```

## Define the Unknown Functions

We have 8 unknown functions with dimensionalities ranging from 2D through 9D.

```python
# The 8 unknown functions (2D through 9D)
factor_counts = [2, 3, 4, 5, 6, 7, 8, 9]
```

## Generate Queries Using Hierarchical Refinement

For each model, we:
1. Start at the centre of the search space (`0.5` in every dimension)
2. Apply a small Gaussian perturbation (`σ = 0.05`) to simulate gradient-informed refinement
3. Clip values to remain within the valid `[0, 1]` range
4. Format the result as a portal query string

```python
print("--- BBO CHALLENGE ROUND 5: ARCHITECTURAL REFINEMENT ---")

for i, count in enumerate(factor_counts):
    center = np.array([0.5] * count)

    # Reducing exploration noise as we build a more complex surrogate
    step_size = 0.05
    refinement = np.random.normal(0, step_size, size=count)

    factors = np.clip(center + refinement, 0, 1)

    # PORTAL FORMAT: x1-x2-x3... to six decimal places
    query_string = "-".join([f"{x:.6f}" for x in factors])
    print(f"Model {i} ({count}D): {query_string}")
```
