# Insurance Pricing Model — Synthetic Factor Generation

This notebook generates synthetic pricing factor values for multiple insurance pricing model versions.  
Each model version uses a different number of **rating factors** (e.g., age, vehicle type, location, claim history, etc.).  

The factor values are drawn from a uniform distribution \([0, 1]\), representing normalised risk scores, exposure metrics, or claim likelihood features.

```python
import numpy as np
```

## Define Model Configurations

Each entry in `factor_counts` represents a pricing model version with that many rating factors:
- **2 factors** → e.g., age & vehicle type  
- **3 factors** → e.g., age, vehicle type, location  
- up to **9 factors** for a highly granular pricing model

```python
factor_counts = [2, 3, 4, 5, 6, 7, 8, 9]
```

## Generate Synthetic Pricing Factors

For each model version, we:
1. Sample `count` values from `Uniform(0, 1)`
2. Sort them for consistency
3. Round to 8 decimal places

```python
pricing_inputs = {}

for i, count in enumerate(factor_counts):
    factors = np.random.uniform(0, 1, size=count)
    factors = np.round(np.sort(factors), 8)
    pricing_inputs[f"pricing_model_version_{i}"] = factors
```

## Display Results

Each model version's factors are printed, separated by ` - ` for easy comparison.

```python
for model_name, factors in pricing_inputs.items():
    formatted_vals = " - ".join(map(str, factors))
    print(f"{model_name}: {formatted_vals}")
```

## Summary Table

Display the results as a structured table using pandas.

```python
import pandas as pd

summary_data = []
for model_name, factors in pricing_inputs.items():
    summary_data.append({
        "Model": model_name,
        "Num Factors": len(factors),
        "Min Factor": factors.min(),
        "Max Factor": factors.max(),
        "Mean Factor": np.round(factors.mean(), 6),
        "Factors": " | ".join(map(str, factors))
    })

df_summary = pd.DataFrame(summary_data)
df_summary
```
