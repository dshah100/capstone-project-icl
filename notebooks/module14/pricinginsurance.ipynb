# BBO Challenge Round 3: SVM-Informed Exploitation

Generate SVM-informed query coordinates for eight unknown insurance pricing functions,
each operating in a different dimensionality (2D through 9D).

**Strategy**: Rather than pure random exploration, this round simulates exploitation
by sampling near a promising centre identified in earlier rounds. A soft-margin SVM
with an RBF kernel was used to classify regions into high- vs. low-performing areas,
and the search radius is tightened around the best-performing region.

```python
import numpy as np

np.random.seed(141)
```

## Define the Exploitation Parameters

Each of the eight unknown functions has a dimensionality ranging from 2 to 9.
We centre the search around `0.5` with a reduced radius of `0.2`, simulating
the refinement of a high-performance region identified by the SVM classifier
from Rounds 1 and 2.

```python
factor_counts = [2, 3, 4, 5, 6, 7, 8, 9]

center = 0.5
search_radius = 0.2
```

## Generate SVM-Informed Queries

For each pricing model version, sample uniformly within the exploitation window
`[centre - radius, centre + radius]` and clip to `[0, 1]`. Output is formatted
as the portal expects: `x1-x2-...-xn` with six decimal places.

```python
print("--- BBO CHALLENGE ROUND 3: SVM-INFORMED QUERIES ---\n")

for i, count in enumerate(factor_counts):
    factors = np.random.uniform(
        center - search_radius,
        center + search_radius,
        size=count
    )
    factors = np.clip(factors, 0, 1)

    query_string = "-".join([f"{x:.6f}" for x in factors])
    print(f"Model {i} ({count}D) Query: {query_string}")
```
