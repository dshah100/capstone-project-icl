# BBO Challenge Round 4: Neural Network & Gradient Strategy

This notebook implements a gradient-based optimization strategy for the Black-Box Optimization (BBO) challenge.

**Approach:** Use a Neural Network surrogate to estimate gradients, starting from the best known region (center of the search space) and stepping along a simulated gradient direction to explore better regions in high-dimensional spaces.

## Setup & Configuration

```python
import numpy as np

# Set seed for Round 4 / Module 16
np.random.seed(161)
```

## Define the Unknown Functions

We have 8 unknown functions with dimensionalities ranging from 2D through 9D.

```python
# The 8 unknown functions (2D through 9D)
factor_counts = [2, 3, 4, 5, 6, 7, 8, 9]
```

## Generate Queries Using Gradient Strategy

For each model, we:
1. Start at the center of the search space (`0.5` in every dimension)
2. Simulate a gradient step: `x_new = x_old + learning_rate * gradient`
3. Clip values to remain within the valid `[0, 1]` range
4. Format the result as a portal query string

```python
print("--- BBO CHALLENGE ROUND 4: NEURAL NETWORK & GRADIENT STRATEGY ---")

for i, count in enumerate(factor_counts):
    center = np.array([0.5] * count)

    learning_rate = 0.1
    simulated_gradient = np.random.uniform(-1, 1, size=count)

    factors = center + (learning_rate * simulated_gradient)

    factors = np.clip(factors, 0, 1)

    # PORTAL FORMAT: x1-x2-x3... each to 6 decimal places
    query_string = "-".join([f"{x:.6f}" for x in factors])

    print(f"Model {i} ({count}D) Query: {query_string}")
```
